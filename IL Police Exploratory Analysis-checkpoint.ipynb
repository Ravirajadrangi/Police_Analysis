```python
# Let's import the libraries we need for this initial analysis:
import pandas as pd
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
import sklearn
# Let's open the Stanford Open Policing Project's CSV file 
# (available here: https://openpolicing.stanford.edu/data/) 
# containing data about all police stops in IL from 2004-2015:
IL = open('IL.csv')
# Let's store the data in a Pandas DataFrame, and take a look at 
# the first five rows (of approximately 4.5 million):
ILdf = pd.read_csv(IL)
# Now we display the first 5 rows using head()
ILdf.head()
```

```python
# There's some weird stuff going on...
# It seems like the county_fips is always either NaN or 17031, so that's useless to us
# Furthermore, at least one stop duration was reported as a negative number, 
# and most stop durations are not reported,
# so this is likely to be useless to us as well

# Let's start our core analysis by removing columns that are not very useful to us
ILdf1 = ILdf.drop(['id','state','county_fips','police_department', 
                   'driver_age_raw', 'violation_raw', 'search_type_raw',
                  'is_arrested','stop_duration','district', 'driver_race_raw',
                  'county_name', 'location_raw'], axis=1)
# Let's take a look at the columns of this new dataframe
list(ILdf1)
```

```python
# We have trimmed down our dataset quite a bit; we now have only 14 columns instead of 26
# Now let's do some preprocessing
# There are some columns that have string datatypes,
# but the data that concerns us within them is purely numerical
# These columns are 'stop_date', 'stop_time', and 'vehicle_type'
# We want to convert 'stop_time' to hours ranging from 0-24,
# 'stop_date' to the day of the year ranging from 1-366,
# and 'vehicle_type' to the vehicle age in years.
# Note that we are not going to consider the brand of the vehicle
# because too many exist and that will greatly increase dimensionality

# Let's start with 'stop_date' and 'vehicle_type':
# Convert the 'stop_date' column into a DatetimeIndex
list(ILdf1)
day_year = pd.DatetimeIndex(ILdf1['stop_date'])
# Convert the 'vehicle_type' column into a year, starting by creating
# a new dataframe instance
ILdf2 = ILdf1
# Then extracting the numeric characters from each string
ILdf2['vehicle_type'] = ILdf2['vehicle_type'].str.extract('(\d+)', expand=False)
# Then converting to integer values
ILdf2['vehicle_type'] = ILdf2['vehicle_type'].astype(float)
# Now let's find the age of the vehicle, write it into 'vehicle_age' and delete 'vehicle_type'
# Note we have a "+ 1" because most vehicles of year N actually release in year N-1
ILdf2['vehicle_age'] = day_year.year - ILdf2['vehicle_type'] + 1
del ILdf2['vehicle_type']
# Now back to the day of the year
# Use the dayofyear method (of the DateTimeIndex class) to get a integer series:
dayofyear = day_year.dayofyear
ILdf2['day_of_year'] = dayofyear
del ILdf2['stop_date']
# Let's check the first five columns to see if the manipulations worked as desired:
ILdf2.head()
```

```python
# Looks good! Now let's do the stop_time:
ILdf2time = pd.DatetimeIndex(ILdf2['stop_time'])
ILdf2time2 = ILdf2time.hour + ILdf2time.minute / 60
ILdf2['stop_time'] = ILdf2time2
# Now let's check to see if the manipulations worked as desired:
ILdf2.head()
```

```python
# Good again (though this computation took quite a while and could be further optimized):
# Now we need to replace the NaNs within the Boolean column "drugs_related_stop" with False
ILdf2['drugs_related_stop'].fillna(False, inplace=True)
# Let's also replace the NaNs in 'search_type' with 'None'
ILdf2['search_type'].fillna('None', inplace=True)
# Checking our work:
ILdf2.head()
```

```python
# There seems to be some messy data in the vehicle_age category, so let's
# filter out the outliers
ILdf2 = ILdf2[ILdf2.vehicle_age >= 0]
ILdf2 = ILdf2[ILdf2.vehicle_age < 70]
```
